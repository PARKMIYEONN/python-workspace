```python
import csv
import matplotlib.pyplot as plt
```

```python
with open('data/ageTotal_202308.csv', 'r') as file:
    data = csv.reader(file)
    header = next(data)   #next()는 제목 인덱스의 리스트 값을 가져온다
    print(header[3], header[3+100], header[106], header[106+100])
    town = '청량리'
    male = list()
    female = list()
    for row in data:
        if town in row[0]:
            for i in range(0,101):
                male.append(int(row[i+3].replace(',','')))
            for j in range(0,101):
                female.append(int(row[j+106].replace(',','')))
print(male)
print(female)
```

```python
# 남성인구수를 음수로 변환
for i in range(len(male)):
    male[i] = -male[i]
print(male)
```

```python
plt.barh(range(101), male)
plt.barh(range(101), female)
```

```python
# 방법2

with open('data/ageTotal_202308.csv', 'r') as file:
    data = csv.reader(file)
    header = next(data)   #next()는 제목 인덱스의 리스트 값을 가져온다
    print(header[3], header[3+100], header[106], header[106+100])
    town = '영등포'
    male = list()
    female = list()
    for row in data:
        if town in row[0]:
            print(row[3:3+100+1])
            male = [int(i.replace(',','')) for i in row[3:104]]
            female = [int(i.replace(',','')) for i in row[106:]]
print(male)
malee = [-i for i in male]
print(malee)
print(female)
```

```python
#시각화
plt.rc('font', family='gulim')   #한글 글씨체 지정해주면 한글로도 가능
plt.title(f'{town} 연령별 남/여 그래프')
plt.barh(range(101), malee, label='남')
plt.barh(range(101), female, label='여')
plt.legend()  #범례 추가
plt.xlabel('인구')
plt.ylabel('연령')
```

```python
# 해당 동의 남여비율 원그래프 시각화
```

```python
male1 = sum(male)
female1 = sum(female)
print(male1, female1)

plt.pie([male1, female1], labels=['남성', '여성'], autopct='%.2f%%', startangle=90)
plt.legend()
```

```python
totalCnt = [male[i] + female[i] for i in range(len(male))]
plt.scatter(male, female, s=totalCnt, c=range(len(male)), alpha=0.5)
plt.colorbar()
plt.xlabel('남자인구수')
plt.ylabel('여자인구수')
```
