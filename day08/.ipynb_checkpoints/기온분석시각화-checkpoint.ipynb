```python
with open('data/seoul-2022.csv', 'r') as file:
    highTemp = []
    while True:
        line = file.readline()
        if line == '':
            break
        templist = line.split(',')
        print(line)
        print(templist[4])
        highTemp.append(templist[4])
        
    print(highTemp)
```

```python
# CSV 활용
with open('data/seoul-2022.csv', 'r') as file:
    data = csv.reader(file)
    highTmeplist = list()
    next(data)   #data에서 첫번째 행은 패스하겠다
    for row in data:
#         print(raw[-1])  #최고기온 찾기, 맨 마지막 데이터를 뽑아낼 때는 -1을 많이 사용하는 편
#         pirnt(f'[{row[-1]}]')
        if len(row) > 1 and row[-1] != '':
            highTmeplist.append(float(row[-1]))
    print(highTmeplist)
    print('최고온도 : ', max(highTmeplist))
    print(len(highTmeplist))
```

## 매년 8월의 최고온도데이터를 시각화

```python
with open('data/seoul-1970and2023.csv', 'r') as file:
    data = csv.reader(file)
    highTmeplist = list()
    next(data)
    for row in data:
        if len(row) > 1 and row[-1] != '':
#         print(row[0].split('-')[1])
            date = row[0].split('-')
            month = date[1]
            if month == '08':
                print(row)
                highTmeplist.append(float(row[-1]))
    print(highTmeplist)
    
plt.figure(figsize=(10,3))
plt.plot(highTmeplist, 'r')
```

```python
# 자신의 생일 기온그래프(최고, 최저) 시각화
with open('data/seoul-1970and2023.csv', 'r') as file:
    data = csv.reader(file)
    highTemplist = list()
    lowTemplist = list()
    years = list()
    next(data)
    for row in data:
        if len(row) > 1 and row[-1] != '':
            date = row[0].split('-')
            year = date[0]
            month = date[1]
            day = date[-1]
            if month == '10' and day == '03':
                print(row)
                highTemplist.append(float(row[-1]))
                lowTemplist.append(float(row[3]))
                years.append(year.strip())
plt.figure(figsize=(15,3))
plt.xticks(rotation=45)    #이 각도대로 틀어져서 보임. 지금은 45도
plt.title('Low&High Temperature on my Birthday')
plt.plot(years, highTemplist, 'r')
plt.plot(years, lowTemplist, 'b')
```
